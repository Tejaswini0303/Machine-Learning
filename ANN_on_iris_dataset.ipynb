```python
import pandas as pd
import numpy as np
import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.utils import to_categorical
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['target']=iris.target.reshape(-1,1)

```

```python
x=df.drop('target',axis=1)
y=to_categorical(df['target'])
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=Sequential()
```

```python
model.add(Dense(10,input_shape=(4,),activation='relu',name='l1'))
model.add(Dense(10,activation='relu',name='l2'))
model.add(Dense(3,activation='softmax',name='l3'))
optimizer=Adam(learning_rate=0.001)
model.compile(optimizer,loss='categorical_crossentropy',metrics=['accuracy'])
```

```python

print(model.summary())
```

```python
model.fit(x_train, y_train, verbose=2, batch_size=10, epochs=200)
```

```python
loss,accuracy=model.evaluate(x_test,y_test)
```

```python
print(loss)
print(accuracy)
```
