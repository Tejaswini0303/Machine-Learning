```python
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
digits=load_digits()
```

```python
df=pd.DataFrame(data=digits.data,columns=digits.feature_names)
```

```python
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df,digits.target,test_size=0.4)
```

```python
model=KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
y_predict
```

```python
accuracy_score(y_test,y_predict)
```

```python
from sklearn.svm import SVC
```

```python
model1=SVC()
```

```python
model1.fit(x_train,y_train)
```

```python
y_predict1=model1.predict(x_test)
```

```python
accuracy_score(y_predict1,y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model2=AdaBoostClassifier(n_estimators=7)
```

```python
model2.fit(x_train,y_train)
```

```python
y_predict2=model2.predict(x_test)
```

```python
accuracy_score(y_predict2,y_test)
```

```python
model3=RandomForestClassifier(n_estimators=3)
```

```python
model3.fit(x_train,y_train)
```

```python
y_predict3=model3.predict(x_test)
```

```python
accuracy_score(y_predict3,y_test)
```
